But we have got the model ready for further use.

```python
def subdraw(ax,filename,max_step=10000):
#     pdb.set_trace()
    arr = np.load(filename)
    steps = list(range(len(arr[0])))
    ax.plot(steps[:max_step], np.log(arr[0])[:max_step],label='training')
#     steps2 = list(range(len(arr[1])))
#     ax.plot(steps2[:max_step]*100, np.log(arr[1])[:max_step],label='validation')
    ax.legend(fontsize=12)
    ax.set_xlabel('step',fontsize=12)
    ax.set_ylabel('RMSE (natrual log)',fontsize=12)

f, (ax1, ax2) = plt.subplots(1, 2, figsize=(15,5))
subdraw(ax=ax1,filename='./img/demo_1_1_pltdata_2019.03.08.15:53:19.npy',max_step=5000)
ax1.set_title('Our model\'s trainning loss',fontsize=15)
subdraw(ax=ax2,filename='./img/demo_1_2_pltdata_2019.03.08.17:18:47.npy',max_step=5000)
ax2.set_title('Target model\'s trainning loss',fontsize=15)
```
